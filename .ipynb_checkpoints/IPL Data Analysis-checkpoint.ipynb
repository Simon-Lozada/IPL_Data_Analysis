## Batsman comparison 

```python
#here we define the variable "balls" 
#We group the column "batsman" with the column "ball", "count" the data, and then convert it to a data frame using the function "reset_index".
balls = df.groupby("batsman")["ball"].count().reset_index()
#here we can see the variable
balls
```

```python
#here we create a variable "runs"
#We group the column "batsman" with the column "batsman_runs", "sum" the data, and then convert it to a data frame using the function "reset_index".
runs = df.groupby("batsman")["batsman_runs"].sum().reset_index()
#here we can see the varible "runs"
runs
```

```python
#here we create a varible "four" this varible is the  our dataframe(df) using the filter "batsman_runs" = 4
four = df[df["batsman_runs"] == 4]
#here we can see the result
four
```

```python
#here wecreate a avarible runs_4
#this varible group the column "batsman" and column "batsman_runs" of our dataframe(four), we "count" the data, then covert it to a dataframe using "reset_index"
runs_4 = four.groupby("batsman")["batsman_runs"].count().reset_index()
#here we put a name to the columns
runs_4.columns = ["batsman", "4s"]
#here we can see the result
runs_4
```

```python
#here we create a varible "six" this varible is the  our dataframe(df) using the filter "batsman_runs" = 6
six = df[df["batsman_runs"] == 6]
#here we can see the result
six
```

```python
#here wecreate a avarible six
#this varible group the column "batsman" and column "batsman_runs" of our dataframe(six), we "count" the data, then covert it to a dataframe using "reset_index"
six = four.groupby("batsman")["batsman_runs"].count().reset_index()
#here we put a name to the columns
six.columns = ["batsman", "6s"]
#here we can see the reusult
six
```

```python
#here we create a new dataframe(player) concat the dataframes "runs", "balls" ,"runs_4" ,"six" then we set axis = 1 
player = pd.concat([runs, balls.iloc[:,1], runs_4.iloc[:,1], six.iloc[:,1]], axis = 1)
#here we can see the result
player
```

```python
#here we use the fillna to convert NaN values to 0
player.fillna(0, inplace = True)
```

```python
#here we can see the results 
player
```

```python
# here we create a new column "striker_rate", "striker_rate" is "batsman_runs" divided by "ball" 
player["striker_rate"] = player["batsman_runs"]/player["ball"]

player
```

```python
#here we create a new variable "grp"
#here we group the columns "match_id", "batsman", "batting_team" of the our dataframe(df), then we "sum" the data "batsman_runs", then we convert it to dataframe using the function "reset_index"
grp = df.groupby(["match_id", "batsman", "batting_team"])["batsman_runs"].sum().reset_index()
#here we can see the result
grp
```

```python
#here we create a new varible "max"
#here we group the columns "batsman" and "batsman_runs" and see the maximun values of the "batsman_runs", then we convert the varible "max" it to dataframe using the function "reset_index" 
max = grp.groupby("batsman")["batsman_runs"].max().reset_index()
#here we put the names of the columns 
max.colums = ["batsman", "max_runs"]
#here we can see the results 
max
```

```python
#here we create a new dataframe concant the "player" dataframe and the "max" dataframe, then we set axis = 1 
player2 = pd.concat([player, max.iloc[:,1]], axis = 1)
#here we can see the result 
player2
```

## Top 10 batsman

```python
#here we create a new data frame by grouping the columns "batsman", "batsman_runs" and sum the data to the "batsman_runs"
max_runs = df.groupby("batsman")["batsman_runs"].sum()
#here we create a graph using "plot(kind = "bar")" and we put the values in ascending order "(ascending = False)" then we we asked for the top 10 "[:10]"
max_runs.sort_values(ascending = False)[:10].plot(kind = "bar")
```

## Top  individual scores batsman

```python
#here we create a new varible agruping the columns "match_id", "batsman", "batting_team" and "sum" the values "batsman_runs", then we convert the varible in dataframe using the function "reset_index", then we asked the values in ascending order 
top_individual = df.groupby(["match_id", "batsman", "batting_team"])["batsman_runs"].sum().reset_index().sort_values(by = "batsman_runs", ascending = False)
#here we see the dataframe using the function "head" and we asked the top 10  
top_individual.head(10)
```
